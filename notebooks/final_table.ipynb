```python
import wandb
from math import isnan 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import wandb
project= "QS_Ours_OnlyCE"
# Project is specified by <entity/project-name>
result_list = []
config_list = []
api = wandb.Api()
runs = api.runs(f"alelab/{project}")
epsilon_ce ={8:0.003921568627, 4:0.1,2:0.7, 1:0.9}
for run in runs:
    if run.state=="finished": 
        config = {}
        config["bit_width_list"] = list(map(int, run.config["bit_width_list"].split(',')))
        config["bit_width"] = int(config["bit_width_list"][0])
        config["epsilon_out"] = run.config["epsilon_out"]
        config["algorithm"] = "Ours (CE)"
        if float(config["epsilon_out"])==epsilon_ce[config["bit_width"]]:
            config["seed"] = run.config["seed"]
            print(run.name)
            #config["epsilonlw"] = run.config["epsilonlw"]
            val = run.history(keys=[f'val_acc_{config["bit_width"]}'], x_axis="epoch")
            test = run.history(keys=[f'test_acc_{config["bit_width"]}'], x_axis="epoch")
            val = val.sort_values(f'val_acc_{config["bit_width"]}')
            best_epoch = val[-1:]["epoch"].values[0]
            val = val[val["epoch"]==best_epoch][f'val_acc_{config["bit_width"]}'].values[0]
            test = test[test["epoch"]==best_epoch][f'test_acc_{config["bit_width"]}'].values[0]
            result_dict = {"accuracy": test}
            result_list.append(result_dict)
            config_list.append(config)
    config_df = pd.DataFrame.from_records(config_list) 
    result_df = pd.DataFrame.from_records(result_list) 
    all_df = pd.concat([config_df,result_df], axis=1)
print("Vanilla")
project= "QSVanilla"
runs = api.runs(f"alelab/{project}")
for run in runs:
    if run.state=="finished":
        print(run.name) 
        config = {}
        config["bit_width_list"] = list(map(int, run.config["bit_width_list"].split(',')))
        config["bit_width"] = int(config["bit_width_list"][0])
        config["algorithm"] = "Baseline"
        config["seed"] = run.config["seed"]
        val = run.history(keys=[f'val_acc_{config["bit_width"]}'], x_axis="epoch")
        test = run.history(keys=[f'test_acc_{config["bit_width"]}'], x_axis="epoch")
        val = val.sort_values(f'val_acc_{config["bit_width"]}')
        best_epoch = val[-1:]["epoch"].values[0]
        val = val[val["epoch"]==best_epoch][f'val_acc_{config["bit_width"]}'].values[0]
        test = test[test["epoch"]==best_epoch][f'test_acc_{config["bit_width"]}'].values[0]
        result_dict = {"accuracy": test}
        result_list.append(result_dict)
        config_list.append(config)

epsilon_l2 ={8:0.1, 4:0.5, 2:0.33333333333, 1:1}
print("L2")
project= "QSL2Abl"
runs = api.runs(f"alelab/{project}")
for run in runs:
    if run.state=="finished": 
        print(run.name)
        config = {}
        config["bit_width_list"] = list(map(int, run.config["bit_width_list"].split(',')))
        config["bit_width"] = int(config["bit_width_list"][0])
        config["epsilonlw"] =  run.config["epsilonlw"]
        config["algorithm"] = "Ours (L2)"
        config["seed"] = run.config["seed"]
        if float(config["epsilonlw"])==epsilon_l2[config["bit_width"]]:
            val = run.history(keys=[f'val_acc_{config["bit_width"]}'], x_axis="epoch")
            test = run.history(keys=[f'test_acc_{config["bit_width"]}'], x_axis="epoch")
            val = val.sort_values(f'val_acc_{config["bit_width"]}')
            best_epoch = val[-1:]["epoch"].values[0]
            val = val[val["epoch"]==best_epoch][f'val_acc_{config["bit_width"]}'].values[0]
            test = test[test["epoch"]==best_epoch][f'test_acc_{config["bit_width"]}'].values[0]
            result_dict = {"accuracy": test}
            result_list.append(result_dict)
            config_list.append(config)
config_df = pd.DataFrame.from_records(config_list) 
result_df = pd.DataFrame.from_records(result_list) 
all_df = pd.concat([config_df,result_df], axis=1)
```

```python
from statistics import NormalDist

def confidence_interval(data, confidence=0.95, std=True):
  if std:
    return np.std(data)
  else:
    dist = NormalDist.from_samples(data)
    z = NormalDist().inv_cdf((1 + confidence) / 2.)
    h = dist.stdev * z / ((len(data) - 1) ** .5)
    return h
```

```python
bitwidths = [1, 2, 4, 8, 32]

acc = {bw:{} for bw in bitwidths}
for bw in bitwidths:
    print("bw: ", bw)
    df_bw = all_df.query(f"bit_width=={bw} and seed<6")
    algorithms = df_bw["algorithm"].unique()
    for algorithm in algorithms:
        mean = df_bw[df_bw["algorithm"]==algorithm]["accuracy"].mean()
        #std = df_bw[df_bw["algorithm"]==algorithm]["accuracy"].std()
        print(algorithm, len(df_bw[df_bw["algorithm"]==algorithm]["accuracy"]))
        std = confidence_interval(df_bw[df_bw["algorithm"]==algorithm]["accuracy"].values, std=True)
        acc[bw][algorithm] = {"mean": mean, "std": std}

            
```
